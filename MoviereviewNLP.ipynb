```python
import time
from threading import *

class hey(Thread):
  def run(self):
    for i in range(5):
      print("hey")
      time.sleep(1)

class hi(Thread):
  def run(self):
    for i in range(5):
      print("hi")
      time.sleep(1)

t1 = hey()
t2 = hi()

t1.start()
time.sleep(0.1)
t2.start()

t1.join()
t2.join()

print("bye")
```
